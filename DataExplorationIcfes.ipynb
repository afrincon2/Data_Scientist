```python
import pandas as pd
import seaborn as sns
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Saber_11__2019-2.csv')
df.head()
```

¿How many columns there are?

```python
df.shape
print("numbers of rows: ", df.shape[0])
print("numbers of columns: ", df.shape[1])
```

The type of each column

```python
print("Type of columns")
for column in df.columns:
    print(column,": ",df[column].dtypes)
```

The different values that a column can take

```python
df.describe()
```

For the non numerical  variables we have

```python
categorical_columns=df.drop(columns=df.describe().columns)
for i in categorical_columns.columns:
    print("For the column ",i, " we can have:")
    a=categorical_columns[i].value_counts()
    print(a)
```

Exploratory analysis:

```python
corr=df.corr ()
# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
analisys=df[(df["FAMI_EDUCACIONPADRE"]=="Secundaria (Bachillerato) completa")|(df["FAMI_EDUCACIONPADRE"]=="Primaria completa")|(df["FAMI_EDUCACIONPADRE"]=="Educación profesional completa")|(df["FAMI_EDUCACIONPADRE"]=="Postgrado")]
analisys
```

```python
plt.rcParams["figure.figsize"] = (5,5)
sns.boxplot( x = 'COLE_BILINGUE',y = 'PUNT_GLOBAL', width=0.5, data = df)
```

```python
plt.rcParams["figure.figsize"] = (5,5)
sns.boxplot( x = 'COLE_NATURALEZA',y = 'PUNT_GLOBAL', width=0.5, data = df)
```

```python
df["FAMI_ESTRATOVIVIENDA"].value_counts()
df.groupby(['FAMI_ESTRATOVIVIENDA']).mean()["PUNT_GLOBAL"]
plt.rcParams["figure.figsize"] = (10,6)
sns.boxplot( x = 'FAMI_ESTRATOVIVIENDA',y = 'PUNT_GLOBAL', width=0.5, data = df)
```

```python
df.groupby(['ESTU_DEPTO_RESIDE'],group_keys=True).mean()["PUNT_GLOBAL"]
#sns.catplot(data=df, x="ESTU_DEPTO_RESIDE", y="PUNT_GLOBAL", jitter=False)
plt.rcParams["figure.figsize"] = (45,10)
sns.boxplot( x = 'ESTU_DEPTO_RESIDE',y = 'PUNT_GLOBAL', width=0.8, data = df)
```

```python
analisys=analisys[["FAMI_EDUCACIONPADRE","PUNT_LECTURA_CRITICA","PUNT_C_NATURALES","PUNT_INGLES","PUNT_MATEMATICAS","PUNT_GLOBAL"]]
sns.pairplot(analisys ,hue="FAMI_EDUCACIONPADRE")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (15,5)
sns.boxplot( x = 'FAMI_EDUCACIONPADRE',y = 'PUNT_GLOBAL', width=0.5, data = analisys)
```
